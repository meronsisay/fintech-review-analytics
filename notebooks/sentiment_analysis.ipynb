### Sentiment Analysis

##### Load your cleaned data

```python
# Load your cleaned data
df = pd.read_csv("../data/processed/reviews.csv")

print(f"Loaded {len(df)} reviews")
print(f"Columns: {df.columns.tolist()}")
df.head()
```

##### Add Sentiment Labels and score

```python
# This one line does everything: loads model, classifies all reviews
df = add_sentiment_to_df(df)

print("Sentiment analysis complete!")
print("\nSentiment distribution:")
print(df['sentiment_label'].value_counts())
```

```python
df.head()
```

##### Aggregate sentiment scores

```python
print("\nBy bank:")
print(aggregate_by_bank(df).round(1))
```

```python
print("\nBy rating:")
for rating, results in aggregate_by_rating(df).items():
    print(f"\n{rating}★:")
    for sentiment, pct in results.items():
        print(f"  {sentiment}: {pct*100:.1f}%")
```
